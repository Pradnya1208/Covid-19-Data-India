```python
Population_URL = "https://api.covid19india.org/v4/min/data.min.json"
state_name = []
state_name_dist = []
st_name = []
state_p = []
```

```python
req_Pop = requests.get(Population_URL)
```

```python
p_data = df(req_Pop.json())
population_data = p_data.T
```

```python
for st in population_data.index:
     if not 'UN' in st:
         state_name.append(st)
         state_name_dist.append(st)

unwanted_st = {'UN', 'Other State', 'Other Region', 'Other', 'Unknown'}
unwanted_st_dist = {'UN', 'TT','Other State', 'Other Region', 'Other', 'Unknown'}
state_name = [ele for ele in state_name if ele not in unwanted_st]
state_name_dist = [ele for ele in state_name_dist if ele not in unwanted_st_dist]
```

```python

for elem in state_name:
    st_name.append(name_correction(elem))
    state_p.append(p_data[elem]['meta']['population'])
    print(name_correction(elem), ":", p_data[elem]['meta']['population'])
    
```

```python
len(state_p)
```

```python
df_state_p = {'State': st_name,
           'Population': state_p}

df_state_pop = pd.DataFrame(df_state_p, columns = ['State', 'Population'])
```

```python
st_name_dist = []
dist_p = []
dist_name = []
```

```python
df_city_p = {'State': st_name_dist,
              'City/Region':dist_name,
           'Population': dist_p}

df_city_pop = pd.DataFrame(df_city_p, columns = ['State', 'City/Region', 'Population'])
```
